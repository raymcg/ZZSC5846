```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense

from keras.utils.np_utils import to_categorical # This does the one hot encoding for us
```

Keras' to_categorical() is typically used for encoding target labels in neural network models due to its simplicity, while scikit-learn's OneHotEncoder is better suited for preprocessing input features, especially when dealing with complex data types or requiring more flexible preprocessing capabilities.

OneHotEncoder from scikit-learn is more flexible and powerful for preprocessing input features, particularly when dealing with string categorical data. It can handle a broader range of data types and is more versatile in terms of handling unseen categories, dealing with missing values, and integrating into data processing pipelines.
OneHotEncoder can encode multiple columns at once and is generally used in preprocessing steps for machine learning models, especially when dealing with mixed data types (numerical and categorical).

# ONE HOT ENCODING OF OUTPUT DATA

```python

# Load Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# One-hot encode the labels
y_one_hot = to_categorical(y) # This is a Keras function
```

```python
print("y before one hot encoding: ")
print(y.shape)
print(y)
print("y after one hot encoding: ")
print(y_one_hot.shape)
print(y_one_hot)
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.3, random_state=42)

# Define the model
model = Sequential()
model.add(Dense(10, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))  # 3 output units for 3 classes

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=10)

# Evaluate the model
_, accuracy = model.evaluate(X_test, y_test)
print(f"Accuracy: {accuracy}")
```

# One hot encoding of an INPUT variable

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
# Simulate some data
np.random.seed(42)
age = np.random.randint(18, 70, 1000).reshape(-1, 1)
income = np.random.randint(30000, 70000, 1000).reshape(-1, 1)
membership = np.random.choice(['Silver', 'Gold', 'Platinum'], 1000).reshape(-1, 1)

# One-hot encode the 'membership' feature
encoder = OneHotEncoder(sparse=False)
membership_encoded = encoder.fit_transform(membership)
```

```python
print("membership before one hot encoding: ")
print(membership.shape)
print(membership[:5])
print("membership after one hot encoding: ")
print(membership_encoded.shape)
print(membership_encoded[:5])
```

```python
# Combine the one-hot encoded features back with the rest of the features
X_final = np.hstack((age, income, membership_encoded))

X_final[:5]
```

```python
# Simulate some target variable (customer satisfaction)
y = np.random.choice([0, 1], 1000)  # Binary outcome for simplicity

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.3, random_state=42)

# Define the model
model = Sequential()
model.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=100, batch_size=10)

# Evaluate the model
_, accuracy = model.evaluate(X_test, y_test)
print(f"Accuracy: {accuracy}")

```
