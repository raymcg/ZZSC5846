# Define Perceptron Class and Methods

```python
class Perceptron(object):

    def __init__(self, no_of_inputs, threshold, learning_rate):
        self.threshold = threshold
        self.learning_rate = learning_rate
        self.weights = np.zeros(no_of_inputs + 1)
           
    def predict(self, inputs):
        # dot product of weights and inputs PLUS bias
        summation = np.dot(inputs, self.weights[1:]) \
                    + self.weights[0] 
        if summation > 0:
          activation = 1
        else:
          activation = 0       
                  
        return activation

    #def class_accuracy(self, training_inputs, labels):
      #try implementing yourself

    def train(self, training_inputs, labels):
        for epoch in range(self.threshold):
            print(f"{epoch=}")

            # Iterate through each row of our training set
            for inputs, label in zip(training_inputs, labels): #zip combines our X values and y labels
                prediction = self.predict(inputs)
                print(f"{prediction=}, {label=}")
                self.weights[1:] += self.learning_rate * (label - prediction) * inputs
                self.weights[0] += self.learning_rate * (label - prediction) #bias
                print(f"{self.weights=}")

            
      #return train_acc, test_acc



```

# Create the Dataset : Inputs and Labels

```python
# Set the dataset inputs : Create our X Matrix
training_inputs = []
training_inputs.append(np.array([1, 1])) # This is row 1 of X (our first training example and it has two features)
training_inputs.append(np.array([1, 0])) # This is row 2 of X (our second training example and it has two features)
training_inputs.append(np.array([0, 1]))
training_inputs.append(np.array([0, 0]))

# Set the dataset class labels
labels = np.array([1, 0, 0, 0]) # this is AND gate example 

# This is our Y vector, or the labels for our training examples.  4 Y values as we have 4 training examples
#labels = np.array([1, 1, 1, 0]) # this is OR gate example 
# Is is saying the answer for row 1 is True, row 2 is True, row 3 is True, row 4 is False

#labels = np.array([1, 0, 0, 1]) # this is XOR gate example
```

# Set Model Parameters

```python
no_of_inputs = 2
threshold=100 # maximum number of training epochs
learning_rate=0.01

# Set the class and train
perceptron = Perceptron(no_of_inputs, threshold, learning_rate)
perceptron.train(training_inputs, labels)
#train_acc, test_acc = perceptron.train(training_inputs, labels)
```

# Explanations

## Program Structure

We are building an X matrix of inputs to train the model: training_inputs (this is a MATRIX, but only of the X Values)
We create a y_vector with the labels for those inputs: labels

We set a number of iterations (threshold) which is how many epochs we cycle through before halting
We set a learning rate (learning_rate) to set how big of jumps we make in our weight tweaks
We initialize our weights as zeros (weights).  This is a single vector with the factors and the bias.  We set all of them to zero intitially.

We initialize the class by creating an instance of the class and having the thresold etc defined in the __init__ of the class
    perceptron = Perceptron(no_of_inputs, threshold, learning_rate)

We then run the function where we train the model
This in in effect just adjusting the weights to arrive at better weights

    perceptron.train(training_inputs, labels)

### How do We Adjust the Weights? How Does the Traing Work?

1.  Create an outer loop equal to how many epochs we want to run 

        for _ in range(self.threshold):

2.  Create an inner loop to go through each row in our training set, row by row (done via the train function)

    First predicting a value given the existing weights (call the prediction function)
    
    Then adjusting the weights based on the results of the prediction (done within the train function)

    We are not using a return statement as we defined a method and so the adjusted weights become a global variables:
         perception.weights

3.  To test our results, we run the predict method with our new test values as our input.

        inputs = np.array([1, 0]) # This is our test set
        output = perceptron.predict(inputs) 

    Note, for our test, it is not telling us whether the result is correct, just what it predicts the result as.
    We could obviously adjust the code if we wanted to, but as written, just the prediction is shown.

    

### Weight adjustment within the Train Function

            for inputs, label in zip(training_inputs, labels): 

                # make a prediction for this training example using the current weights
                ## This is where we run the prediction function we created
                prediction = self.predict(inputs)  

                # Adjust all the non-bais weights
                self.weights[1:] += self.learning_rate * (label - prediction) * inputs

                # Adjust the bias weight
                self.weights[0] += self.learning_rate * (label - prediction) #bias
        
        So our adjustment for the non-bias is:
            Learning Rate * (actual answer - prediction) * our vector of X values

        Here is an example of a row that would cause the weights to adjust:
            prediction=0, label=1
            inputs=array([1, 1])
            Weights before: weights=array([0., 0., 0.])
            Weights after: weights=array([0.01, 0.01, 0.01])
